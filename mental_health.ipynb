## Importing ML Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Read the csv

```python
dataset1 = pd.read_csv('Student Mental health.csv')
dataset = pd.read_csv('Student_Mental_health.csv')
```

```python
dataset['cgpa'] = dataset1['What is your CGPA?']
```

```python
dataset
```

```python
dataset.drop("course", axis='columns', inplace=True)
```

```python
dataset
```

```python
dataset['Depression'] = dataset1['Do you have Depression?']
dataset['Anxiety'] = dataset1['Do you have Anxiety?']
dataset['Panic'] = dataset1['Do you have Panic attack?']
```

```python
dataset
```

```python
drop_cols = ['Depression score','Anxiety Score','ADHD score']

dataset.drop(drop_cols, axis='columns', inplace=True)
```

```python
dataset
```

```python
def convert_cgpa_range_to_midpoint(cgpa_range):
    range_values = cgpa_range.split('-')
    start_value = float(range_values[0])
    end_value = float(range_values[1])
    mid_point = ((start_value + end_value)/2)*2.5
    return mid_point

dataset['cgpa'] = dataset['cgpa'].apply(convert_cgpa_range_to_midpoint)
```

```python
dataset
```

```python
df = dataset
df
```

```python
df
```

```python
def changeyear(year):
    year_val = int(year[-1])
    return year_val

df['Your current year of Study'] = df['Your current year of Study'].apply(changeyear)
```

```python
df
```

```python
def yesno(val):
    if val == "Yes":
        return 1
    elif val == "No":
        return 0

df['Depression'] = df['Depression'].apply(yesno)
df['Anxiety'] = df['Anxiety'].apply(yesno)
df['Panic'] = df['Panic'].apply(yesno)
```

```python
def salarychange(salary):
    changed_salary = pow(salary,1.2)
    changed_salary = pow(changed_salary,1.4)
    return changed_salary

df['What is your Family income?'] = df['What is your Family income?'].apply(salarychange)

df
```

```python
def changesalary(salary):
    changed_salary = salary/80
    changed_salary = changed_salary*100000
    return changed_salary

df['What is your Family income?'] = df['What is your Family income?'].apply(changesalary)

df
```

## Feature Scaling

```python
dataframe = df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df = scaler.fit_transform(df)
```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
df_train1,df_test1,depression_score_train,depression_score_test =  train_test_split(df, depression_score, test_size=0.2, random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

## Model to test Depression

```python
model1 = RandomForestClassifier()
model3 = RandomForestClassifier()
model2 = LogisticRegression()
```

```python
voting_classifier_1 = VotingClassifier(estimators=[
    ('rf', model1),
    ('lr', model2),
    ('rf2', model3)
], voting='soft')
```

```python
voting_classifier_1.fit(df_train1, depression_score_train)
```

```python
voting_classifier_1.score(df_test1, depression_score_test)
```

## Model to test Anxiety
